## Top College Majors

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')
# ages <28
# https://github.com/fivethirtyeight/data/tree/master/college-majors
```

#### Total Majors by Job Type

```python
college = df[['College_jobs', 'Non_college_jobs', 'Low_wage_jobs']]
college.sum().plot(kind='bar')
college.sum()
```

#### Top 5 Majors by Total

```python
top5_majors_by_total = df[['Major', 'Total']].value_counts().sort_values()[:5]
top5_majors_by_total
```

```python
top5_majors_by_total.plot.pie(figsize=(8,8))
```

```python
# plot
fig, ax = plt.subplots()

ax.bar(top5_majors_by_total.index[0][0], top5_majors_by_total.index[0][1])
ax.bar(top5_majors_by_total.index[1][0], top5_majors_by_total.index[1][1])
ax.bar(top5_majors_by_total.index[2][0], top5_majors_by_total.index[2][1])
ax.bar(top5_majors_by_total.index[3][0], top5_majors_by_total.index[3][1])
ax.bar(top5_majors_by_total.index[4][0], top5_majors_by_total.index[4][1])

plt.show()
```
